```python
import cv2
import numpy as np
import os
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename),cv2.IMREAD_COLOR)
        if img is not None:
            images.append(img)
    return images
output=load_images_from_folder(r"C:\Users\vrajs\Desktop\SEM 7\FINAL CODE YOLO\pothole-detection-100.v4-3rddataset.yolov5pytorch\test\images")
print(output)
```
